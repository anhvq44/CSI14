## Import các thư viện

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

## Thêm dữ liệu + xử lí dữ liệu thô

```python
df = pd.read_csv('temperatures.csv')
df.head(5)
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates()
```

```python
df.info()
```

```python
# chia input/ output
f_input = df['Farenheit'].to_numpy()
c_output = df['Celcius'].to_numpy()
type(f_input), type(c_output)
```

## Huấn luyện mô hình

```python
#tao random
keras.utils.set_random_seed(42)
```

```python
#khai bao mo hinh
model = keras.Sequential([
    keras.layers.Dense(units=10, input_shape=[1]),
    keras.layers.Dense(units=1) #output layer
])

model.compile(loss="mean_squared_error", optimizer='adam')
model.summary()
```

```python
model.fit(f_input, c_output, epochs=100) #huan luyen mo hinh
```

## Dự đoán

```python
f_user_input = np.array([90])
#thuc te
c_output_real = round((f_user_input[0]-32)/1.8, 2)
#prediction
c_output_pred = model.predict(f_user_input)
print(f"Real output: {c_output_real} degree celsius")
print(f"Prediction output: {c_output_pred[0][0]:.2f} degree celsius")
```
