```python
import os
import pandas as pd
from dotenv import load_dotenv
import google.generativeai as genai
```

```python
from IPython.display import Markdown
```

```python
load_dotenv()
google_api_key = os.getenv("GOOGLE_API_KEY")
genai.configure(api_key=google_api_key)
```

```python
MODEL_VERSION = "gemini-2.5-flash"
model = genai.GenerativeModel(MODEL_VERSION,
                              system_instruction="""
                              Bạn là một trợ lý AI có nhiệm vụ hỗ trợ cung cấp menu cho nhà hàng. Menu gồm nhứng món sau:
                              1. Pizza New York
                              2. Salad
                              3. Tomato Spaghetti
                              4. Tiramisu
                              Cho mỗi món ăn, bạn cần đưa ra kết quả theo format <tên món> - <mô tả món>. Cho mô tả món, hãy chú trọng miêu tả hương vị đặc trưng, thơm ngon, và thảnh phần chính của nó, trong khoảng 1-2 câu.
                              """)
```

```python
prompt = "Menu"
response = model.generate_content(prompt)
Markdown(response.text)
```
